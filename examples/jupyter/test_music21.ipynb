```python
from music21 import *
from ziffers import *

twinkle = "q [: 0 0 4 4 5 5 h4 q 3 3 2 2 1 1 h0 <(: q4 4 3 3 2 2 h1 :) r> :]"

parsed = zparse(twinkle, key='g4', scale='Major')
score = to_music21(parsed, time="4/4", bpm=120)

score.show("midi")
score.show()
```
